```python
# Open
# Import Libraries
import cv2
from google.colab.patches import cv2_imshow

# Reading image using open cv
Image = cv2.imread('brain.jpg')

# converting RGB image to gray
Grayscale = cv2.cvtColor(Image, cv2.COLOR_RGB2GRAY)

# Displaying image
cv2_imshow(Image)
cv2_imshow(Grayscale)
```

```python
# Matplotlib & Numpy
# Import Libraries

import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np

# Reading image using matplotlib
Image = img.imread('brain.jpg')

# Grayscale operations using numpy
grayscale = np.dot(Image[..., :3], [0.299, 0.587, 0.114])

# displaying image
plt.subplot(1, 2, 1)
plt.imshow(Image)
plt.subplot(1, 2, 2)
plt.imshow(grayscale, cmap = 'gray')
```

```python
# Sckitimage & Matplotlib
# Import Libraries
import matplotlib.pyplot as plt
import matplotlib.image as img
from skimage.color import rgb2gray

# Read the images
Image = img.imread('brain.jpg')

# Converting rgb to gray using skimage libraries
grayscale = rgb2gray(Image)

# Displaying the grayscale image
plt.subplot(1, 2, 1)
plt.imshow(Image)
plt.subplot(1, 2, 2)
plt.imshow(grayscale, cmap = 'gray')
```

```python
# PIL/Pillow
# Import Libraries
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Read the image
Image = Image.open('brain.jpg')

# convert the image into grayscale using PIL
grayscale = Image.convert('L')

# preparing data for matplot display
grayscale = np.array(grayscale)

# Displaying the image
plt.subplot(1, 2, 1)
plt.imshow(Image)
plt.subplot(1, 2, 2)
plt.imshow(grayscale, cmap = 'gray')

```

**Notes:**

1. For **open CV** the conversion of RGB to gray is very easy which is

**cv2.cvtColor**

2. For **Matplotlib** we need to use either numpy or scikit image. Because Matplotlib helps only in displaying image it doen't handle mathematical operations.

The numpy converts RGB to gray using the concept of **weighted average**. Which takes the average of R G and B values converts 3 channels to single channel. So numpy has a special mixer called **np.dot**. And this time while displaying image using matplotlib we need to mention cmap='gray' if we don't mention then they think its a kind of color image so they just gives the random color image of the average values calcuated.

3. The **scikit image** conversion is easy since they have inbuilt function called **rgb2gray** which converts the color image to graycscale and on top of that we have to use matplotlib to display the images.

4. For **pillow** its again easy we just need to use **convert('L')**
But in pillow it sees the image as object not as array. so we need to import numpy to convert them to array for displaying them using matplotlib.
