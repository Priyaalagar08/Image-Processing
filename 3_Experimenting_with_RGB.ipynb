**Experiment:** Displaying R, G , B channels seperately

**Method 1**

Using open cv and matplotlib to perform this.

Open CV uses BGR

PIL, Sk image, Matplotlib, tensorflow, pytorch uses RGB

```python
# Import Libraries
import cv2
import matplotlib.pyplot as plt

# Reading image
Image = cv2.imread('brain.jpg')

#Converting image to grayscale
Gray_Image = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)

# Displaying RGB image seperately from grayscale
Red, Green, Blue = cv2.split(Image)

# Showing the image
plt.subplot(231)
plt.imshow(Image)
plt.title('Original Image')

plt.subplot(232)
plt.imshow(Gray_Image, cmap= 'gray')
plt.title('Grayscale Image')

plt.subplot(233)
plt.imshow(Red)
plt.title('Red Channel')

plt.subplot(234)
plt.imshow(Green)
plt.title('Green Channel')

plt.subplot(235)
plt.imshow(Blue)
plt.title('Blue Channel')
```

**Method 2**

Using PIL, Numpy and Matplotlib libraries

```python
# Import libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Reading image
Image = Image.open('brain.jpg')

# Converting Image to grayscale
Gray_Scale = Image.convert('L')

# converting the image into arrays
Image = np.array(Image)
Gray_Image = np.array(Gray_Scale)

# Splitting r g b using numpy
r = Image[:,:,0]
g = Image[:,:,1]
b = Image[:,:,2]

# Showing the image
plt.subplot(231)
plt.imshow(Image)
plt.title('Original Image')

plt.subplot(232)
plt.imshow(Gray_Image, cmap= 'gray')
plt.title('Gray_Image')

plt.subplot(233)
plt.imshow(r)
plt.title('Red Channel')

plt.subplot(234)
plt.imshow(g)
plt.title('Green Channel')

plt.subplot(235)
plt.imshow(b)
plt.title('Blue Channel')
```

**Using Scikit Libraries**

Here we have only one option which is using Numpy libraries because scikit dont't have anyother options like split

```python
# Import libraries
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np

# Reading image
Image = io.imread('brain.jpg')

# Grayscale conversion
gray_image = rgb2gray(Image)

# Splitting using numpy
r = Image[:,:,0]
g = Image[:,:,1]
b = Image[:,:,2]

# Displaying the image
plt.subplot(231)
plt.imshow(Image)
plt.title('Original Image')

plt.subplot(232)
plt.imshow(gray_image, cmap= 'gray')
plt.title('Gray_Image')

plt.subplot(233)
plt.imshow(r)
plt.title('Red Channel')

plt.subplot(234)
plt.imshow(g)
plt.title('Green Channel')

plt.subplot(235)
plt.imshow(b)
plt.title('Blue Channel')
```
