**Simple Method**

```python
# Import Libraries
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Reading the image
image = cv2.imread('brain.jpg')

# Resizing the image
image_resize1 = cv2.resize(image, (500, 500))       # actual image size is 290*174 and we are asking to resize them into 500 * 500
image_resize2 = cv2.resize(image, (image.shape[1]//2,image.shape[0]//2)) # if we don't know the shape of the actual image and hence  we ask them to resize half the shape
# [1] is the width and [0] is the height, we divide them by 2 so we get half size of the image, if we need quater size we divide by 4

# plot the image
plt.subplot(1,2,1)
plt.imshow(image_resize1)
plt.title("Resize 1")
plt.subplot(1,2,2)
plt.imshow(image_resize2)
plt.title("Resize 2")

# print shape
print(image_resize1.shape)
print(image_resize2.shape)
```

```python
# Import Libraries
import cv2
import matplotlib.pyplot as plt

# Reading Images
Original_Image = cv2.imread('brain.jpg')

# Use cv2.resize for scaling and interpolation
# dsize=None indicates that output size will be calculated from fx and fy
# if we are using interpolation then there should be a target size called dsize (some fixed size) if we dont give them then provide as none and set the scaling factor fx and fy which is for width and height
Test_image1 = cv2.resize(Original_Image, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_AREA)
Test_image2 =cv2.resize(Original_Image, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC)
Test_image3 = cv2.resize(Original_Image, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_LINEAR)
Test_image4 = cv2.resize(Original_Image, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_NEAREST)

# Plotting image
plt.subplot(2,3,1)
plt.imshow(Original_Image)
plt.title("Original Image")

plt.subplot(2,3,2)
plt.imshow(Test_image1)
plt.title("INTER_AREA")

plt.subplot(2,3,3)
plt.imshow(Test_image2)
plt.title("INTER_CUBIC")

plt.subplot(2,3,4)
plt.imshow(Test_image3)
plt.title("INTER_LINEAR")

plt.subplot(2,3,5)
plt.imshow(Test_image4)
plt.title("INTER_NEAREST")
```
