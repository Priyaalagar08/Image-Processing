```python
# Loading and displaying images using open CV
# CV2.imread and cv2.imshow is used, since I am using google collab it doesn't support cv2.imshow so I am going to import cv2_show from google.colab

# Import Libraries
import cv2
from google.colab.patches import cv2_imshow

# Reading Image (I have already uploaded image in collab space)
brain = cv2.imread('brain.jpg')

# Displaying image
cv2_imshow(brain)

# Helping in retaining the window, again since I sue google collab it doesn't have any effect. Can be seen in other python platforms
cv2.waitKey(0)
cv2.destroyAllWindows()

# prints the matrix of 3 channels (RGB)
print(brain)
```

```python
# Loading and displaying images using Matplotlib
# imread and imshow is used

# Import required libraries
import matplotlib.pyplot as plt
import matplotlib.image as img

# Reading image from the collab space
brain = img.imread('brain.jpg')

# Displaying image
plt.imshow(brain)

```

```python
# Loading and displaying images using scikit image
# Either we can use our own data or already existing data(scikit image already have their own data)
# Ifown data is used then import io
# if predefined data is used then import data

# Own data
# Import Libraries
from skimage import io

# Read the image
brain = io.imread('brain.jpg')

# Display the image
io.imshow(brain)
```

```python
# Using predefined data by skimage
from skimage import data, color
import matplotlib.pyplot as plt

# Read the image
brain = data.brain()

# display the image
#io.imshow(brain[0]) # io.show doen't helps in displaying color image hence we use matplot libto display color image
plt.imshow(brain[0])
plt.title('Original RGB image')
plt.show()
```
